```python
# ====== 1. Instalación de librerías (opcional en Colab) ======
!pip install pandas numpy matplotlib scikit-learn --quiet

# ====== 2. Importar librerías ======
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ====== 3. Crear dataset de ejemplo ======
np.random.seed(42)
dias = np.arange(1, 31)
precios = 100 + np.random.randn(30).cumsum()
data = pd.DataFrame({'Dia': dias, 'Precio': precios})
data.head()
```

```python
def entrenar_modelo(df):
  """función dummy de modelo"""
  X = df[['Dia']]
  y = df['Precio']
  modelo = LinearRegression()
  modelo.fit(X, y)
  return modelo

modelo = entrenar_modelo(data)
```

```python
#predicción y visualización

dias_futuros = np.arange(31, 41).reshape(-1,1)
predicciones = modelo.predict(dias_futuros)

plt.figure(figsize=(10,5))
plt.plot(data['Dia'], data['Precio'], label='Precios reales', marker='o')
plt.plot(dias_futuros, predicciones, label='Predicciones', marker='x')
plt.xlabel('Día')
plt.ylabel('Precio')
plt.title('Predicción de precios de ejemplo')
plt.legend()
plt.show()
```
